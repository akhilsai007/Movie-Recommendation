```python
# Assuming you have a DataFrame called 'movies' with a 'genres' column

def clean_genres(genres):
    # Split the genres string into a list of individual genres
    genre_list = genres.split('|')
    # Take the first genre as the main genre
    main_genre = genre_list[0]
    return main_genre

# Apply the clean_genres function to update the 'cleaned_genres' column
movies['cleaned_genres'] = movies['genres'].apply(clean_genres)

# Display the value counts of the cleaned genres
print(movies['cleaned_genres'].value_counts())
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from collections import defaultdict
from prettytable import PrettyTable
from surprise import accuracy
from surprise.model_selection import KFold
from surprise.model_selection import cross_validate
from surprise.model_selection import LeaveOneOut
from math import log
```

```python
# Load the MovieLens 1M dataset
data_info_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
data = pd.read_csv(r'C:\Users\syamsaiakhil.d\OneDrive - Qentelli\Desktop\Movie Recommendation Project\ml-1m dataset\ratings.dat', sep='::', names=data_info_cols, encoding='latin-1', engine='python')
# Define a Reader object to parse the data
reader = Reader(rating_scale=(1, 5))
# Load the dataset into Surprise's data structure
dataset = Dataset.load_from_df(data[['user_id', 'movie_id', 'rating']], reader)
movie_info_cols = ['movie_id', 'title', 'genres']
movies = pd.read_csv(r'C:\Users\syamsaiakhil.d\OneDrive - Qentelli\Desktop\Movie Recommendation Project\ml-1m dataset\movies.dat', sep='::', names=movie_info_cols, encoding='latin-1', engine='python')
```

```python
# Split the data into training and testing sets
trainset, testset = train_test_split(dataset, test_size=0.2)
```

```python
# Create and train an SVD model
model = SVD()
model.fit(trainset)
```

```python
# Use the model to make predictions on the testset
predictions = model.test(testset)
# Calculate RMSE (Root Mean Squared Error)
rmse = accuracy.rmse(predictions)
print(f"RMSE: {rmse:.2f}")
# Calculate MAE (Mean Absolute Error)
mae = accuracy.mae(predictions)
print(f"MAE: {mae:.2f}")
# Define the value of k for precision@k and recall@k
k_values = [3, 5]
# Initialize a KFold cross-validation
kf = KFold(n_splits=5)
# Perform 5-fold cross-validation to evaluate RMSE and MAE
cross_val_results = cross_validate(model, dataset, measures=['RMSE', 'MAE'], cv=kf, verbose=True)
# Calculate Precision, Recall, MAP, and MRR for each k value
for k in k_values:
    precision = 0
    recall = 0
    map_at_k = 0
    mrr = 0
    for trainset, testset in kf.split(dataset):
        model.fit(trainset)
        predictions = model.test(testset)
        top_n = defaultdict(list)
        for uid, iid, true_r, est, _ in predictions:
            top_n[uid].append((iid, est))
        for uid, user_ratings in top_n.items():
            user_ratings.sort(key=lambda x: x[1], reverse=True)
            n_rel = sum((true_r >= 4) for (_, true_r) in user_ratings)
            n_rec_k = sum((est >= 4) for (_, est) in user_ratings[:k])
            precision += n_rec_k / k
            recall += n_rec_k / n_rel if n_rel != 0 else 0
            # Calculate Average Precision (AP)
            ap = 0
            num_hits = 0
            for i, (item, rating) in enumerate(user_ratings):
                if rating >= 4:
                    num_hits += 1
                    ap += num_hits / (i + 1)
            if num_hits > 0:
                map_at_k += ap / num_hits
            # Calculate MRR
            if n_rec_k > 0:
                rank = next(i for i, (item, rating) in enumerate(user_ratings) if rating >= 4) + 1
                mrr += 1 / rank
    precision /= len(testset)
    recall /= len(testset)
    map_at_k /= len(testset)
    mrr /= len(testset)
    print(f'Precision@{k}: {precision:.4f}')
    print(f'Recall@{k}: {recall:.4f}')
    print(f'MAP@{k}: {map_at_k:.4f}')
    print(f'MRR@{k}: {mrr:.4f}')

expected_rr = 0
for uid, user_ratings in top_n.items():
    rank = 1  # Initialize rank
    relevance = 0  # Initialize relevance
    for _, est in user_ratings:
        if relevance < 1:
            err = (2 ** relevance - 1) / (2 ** 1)
            expected_rr += err / rank
        rank += 1
expected_rr /= len(top_n)
print(f'Expected Reciprocal Rank (ERR): {expected_rr:.4f}')
# Calculate Normalized Discounted Cumulative Gain (NDCG)
ndcg = 0
for uid, user_ratings in top_n.items():
    dcg = 0
    idcg = 0
    for i, (_, est) in enumerate(user_ratings[:k]):
        relevance = 1 if est >= 4 else 0  # Assuming relevance is 1 if est >= 4, else 0
        dcg += (2 ** relevance - 1) / log(i + 2, 2)
        idcg += (2 ** 1 - 1) / log(i + 2, 2)
    ndcg += dcg / idcg
ndcg /= len(top_n)
print(f'Normalized Discounted Cumulative Gain (NDCG@{k}): {ndcg:.4f}')
```
